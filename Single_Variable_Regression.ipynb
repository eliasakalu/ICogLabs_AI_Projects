```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Input the data
x=np.array([1,2,3,4,5])
y=np.array([50,55,65,70,80])
```

```python
#Visualize
plt.scatter(x,y,label="the data distribution",color="red")
plt.xlabel("Study Hour")
plt.ylabel("Score")
plt.title("Score vs Study Hour")
plt.plot()
```

```python
#Inital Guess
w=0
b=0
alpha=0.1
iteration=1000
```

```python
def cost(x,y,w,b):
  m=x.shape[0]
  y_pred=w*x+b
  cost=np.sum((y_pred-y)**2)
  cost/=(2*m)
  return cost
def gradient_descent(x,y,w,b,a,iter):
  m=x.shape[0]
  for i in range(iter):
    y_pred=w*x+b

    dj_dw=(1/m)*np.matmul(y_pred-y,x)
    dj_db=(1/m)*np.sum(y_pred-y)

    w=w-a*dj_dw
    b=b-a*dj_db
  return w,b
```

```python
w_final,b_final=gradient_descent(x,y,w,b,alpha,iteration)
print(f"final adjusted w={w_final},b={b_final} with cost={cost(x,y,w_final,b_final)}")
```

```python
#Data vs Our Model
plt.scatter(x,y,label="Actual Data",color="red")
plt.plot(x,w_final*x+b_final,label="Model",color="blue")
plt.xlabel("Study Hour")
plt.ylabel("Score")
plt.title("Score vs Study Hour")
plt.plot()
```

```python
#inference
def predict(x,w_final,b_final):
  y_pred=w_final*x+b_final
  return y_pred
```

```python
#Application
input=float(input("Enter the study Hour:"))
output=predict(input,w_final,b_final)
print(f"The Score is {output:.2f}")
```
