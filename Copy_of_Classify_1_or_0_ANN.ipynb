```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Data for training and testing
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.catvsdogs.load()

#Filter the data for our binary classification
train_mask=np.where((y_train==0)|(y_train==1))
test_mask=np.where((y_test==0)|(y_test==1))

x_train,y_train=x_train[train_mask],y_train[train_mask]
x_test,y_test=x_test[test_mask],y_test[test_mask]

#Change to vector from matrix
x_train,x_test=x_train.reshape(-1,28*28),x_test.reshape(-1,28*28)
print(y_test[:3])
```

```python
#Model Definingg
model=tf.keras.Sequential([
    tf.keras.layers.Dense(units=30,activation="relu"),
    tf.keras.layers.Dense(units=15,activation="relu"),
    tf.keras.layers.Dense(units=1,activation="sigmoid")

])

model.compile(optimizer="adam",loss=tf.keras.losses.BinaryCrossentropy(),metrics=["accuracy"])
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10)
```

```python
predicted_class=model.predict(x_test[2].reshape(-1,28*28))
print(f"The model recognizes the number {int(predicted_class[0][0])}")
```
